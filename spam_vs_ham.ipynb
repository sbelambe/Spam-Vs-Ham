```python
df2 = pd.read_csv('/content/drive/MyDrive/SpamVsHam/email_classification.csv')
df2.head()
```

email.csv: https://www.kaggle.com/datasets/ashfakyeafi/spam-email-classification

```python
df3 = pd.read_csv('/content/drive/MyDrive/SpamVsHam/email.csv')
df3.head()
```

```python
# Combining the dataframes
# Label and Message

#first convert df1 1 and 0 labels to ham and spam
df1['label'].replace(1, 'spam', inplace=True)
df1['label'].replace(0, 'ham', inplace=True)
df1.head()

#combine dataframes now
combined_df = pd.DataFrame()

col1_label = df1['label']
col2_label = df2['label']
col3_label = df3['Category']
combined_series_label = pd.concat([col1_label, col2_label, col3_label])

col1_message = df1['text']
col2_message = df2['label']
col3_message = df3['Message']
combined_series_message = pd.concat([col1_message, col2_message, col3_message])

combined_df = pd.DataFrame({'label': combined_series_label, 'message': combined_series_message})
combined_df.size
```

```python
combined_df.head()
```

# random forest classification
no prior feature selection done before this

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
from sklearn.model_selection import train_test_split

#features = x, target = y
X = combined_df.drop('label', axis=1)
y = combined_df['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

rf = RandomForestClassifier()

rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:",  accuracy)
```
